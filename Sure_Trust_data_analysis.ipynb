```python
import pandas as pd

# Replace 'your_excel_file.xlsx' with the actual path to your Excel file
excel_file = '/content/retail_dataset.xlsx'

try:
    df = pd.read_excel(excel_file)
    display(df.head())
except FileNotFoundError:
    print(f"Error: The file '{excel_file}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# 1. Data Overview
print("Data Overview:")
display(df.head())
display(df.info())
```

```python
# 2. Descriptive Statistics
print("\nDescriptive Statistics:")
display(df.describe())
```

```python
# 3. Categorical Data Analysis
print("\nCategorical Data Analysis:")
categorical_cols = ['Channel', 'City', 'Category', 'Brand', 'Subcategory', 'Customer_Segment']
for col in categorical_cols:
    print(f"\n{col} Distribution:")
    display(df[col].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 4. Order Value Analysis
print("\nOrder Value Analysis:")
plt.figure(figsize=(10, 6))
sns.histplot(df['Order_Value'], bins=50, kde=True)
plt.title('Distribution of Order Value')
plt.xlabel('Order Value')
plt.ylabel('Frequency')
plt.show()

# 5. Churn and Fraud Analysis
print("\nChurn and Fraud Analysis:")
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.countplot(x='Churn_30D', data=df)
plt.title('Churn Distribution (30 Days)')
plt.xlabel('Churn (1=Yes, 0=No)')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
sns.countplot(x='Fraud_Label', data=df)
plt.title('Fraud Distribution')
plt.xlabel('Fraud Label')
plt.ylabel('Count')

plt.tight_layout()
plt.show()

# 6. Relationship Analysis (Example: Channel vs Order Value)
print("\nRelationship Analysis (Channel vs Order Value):")
plt.figure(figsize=(10, 6))
sns.boxplot(x='Channel', y='Order_Value', data=df)
plt.title('Order Value by Channel')
plt.xlabel('Channel')
plt.ylabel('Order Value')
plt.show()

# 7. Finish task (Summary will be provided in a text response after code execution)
```
